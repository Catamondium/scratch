# Playing around

```python
from IPython.display import Math
from sympy import *
XS = range(0, 1000)
x = Symbol('x')
f = 2*x + 5
f_real = lambdify(x, f, 'numpy')
f
```

```python
from matplotlib.pyplot import plot, show
ys = [f_real(x) for x in XS]
%matplotlib inline
plot(XS, ys)
show()
```

```python
%%latex
\begin{align}
\int_{-\infty}^{+\infty} 2x + 5 \ dx = x^2 + 5x + C
\end{align}
```

```python
from matplotlib.pyplot import plot, show
fint = integrate(f)
fint_real = lambdify(x, fint, 'numpy')
ys = [fint_real(x) for x in XS]
%matplotlib inline
plot(XS, ys)
show()
```
